```python
# Treat text columns
df.drop("Name",axis=1,inplace=True)
```

```python
df.head()
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
#Seprate input and output columns
x=df.drop("Survived",axis=1)
y=df[["Survived"]]
```

```python
x.head()
```

```python
y.head()
```

```python
# Train machine learning model
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.24, random_state=42)
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
y_test.head()
```

```python
y_train.head()
```

```python
# import knn alogorithm
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier(n_neighbors=5)
```

```python
KNN.fit(x_train,y_train)
```

```python
ypread=KNN.predict(x_test)
```

```python
ypread
```

```python
from sklearn.metrics import accuracy_score,plot_confusion_matrix,classification_report
```

```python
accuracy_score(y_test,ypread)
```

```python
plot_confusion_matrix(KNN,x_test,y_test)
```

```python
classification_report(y_test,ypread)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x_train_s=scalar.fit_transform(x_train)
```

```python
x_test_s=scalar.transform(x_test)
```

```python
ypred=KNN.predict(x_test_s)
```

```python
ypred3=DT1.predict(x_test)
```

```python
ypred3
```

```python
accuracy_score(y_test,ypred3)
```

```python
ypredtrain1=DT1.predict(x_train)
```
